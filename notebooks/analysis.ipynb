## Content Type Distribution

### Movies v/s TV Shows

```python
plt.figure(figsize = (8, 6))
type_counts = df['type'].value_counts()
colors = ['#E50914', '#564d4d']

type_counts.plot(kind = 'bar', color = colors, edgecolor = 'black', linewidth = 1.2)
plt.title('Netflix Content Distribution: Movies v/s TV Shows', fontsize = 16, fontweight = 'bold', pad = 20)
plt.xlabel('Content Type', fontsize = 12, fontweight = 'bold')
plt.ylabel('Number of Titles', fontsize = 12, fontweight = 'bold')
plt.xticks(rotation = 0, fontsize = 11)
plt.yticks(fontsize = 11)

for i, v in enumerate(type_counts):
    plt.text(i, v + 50, str(v), ha = 'center', fontweight = 'bold', fontsize = 12)
plt.tight_layout()
plt.savefig(r'C:\Users\Administrator\Desktop\ProjectDataDashboard\visualizations\viz1_content_type.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
movies_pct = (type_counts['Movie'] / type_counts.sum()) * 100
tv_show_pct = (type_counts['TV Show'] / type_counts.sum()) * 100
print(f"\n INSIGHT: {type_counts['Movie']:,} Movies ({movies_pct:.1f}%) vs {type_counts['TV Show']:,} TV shows ({tv_show_pct:.1f}%)")
print(f"Netflix has a strong preference for {'Movies' if movies_pct > 50 else 'TV shows'}")
```

## Top 10 Content Producing Countries

```python
countries_split = df['country'].dropna().str.split(', ')
all_countries = []
for country_list in countries_split:
    all_countries.extend(country_list)
```

```python
country_counts = pd.Series(all_countries).value_counts().head(10)
country_counts
```

```python
plt.figure(figsize=(12, 7))
country_counts.plot(kind = "barh", color = '#E50914', edgecolor = 'black', linewidth = 1.2)
plt.title("Top 10 Content Producing Countries on Netflix", fontsize = 16, fontweight = 'bold', pad = 20)
plt.xlabel("Number of Titles", fontsize = 12, fontweight = 'bold')
plt.ylabel("Country", fontsize = 12, fontweight = 'bold')
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)
for i, v in enumerate(country_counts):
    plt.text(v + 20, i, str(v), va='center', fontweight='bold', fontsize=10)

plt.tight_layout()
plt.savefig(r'C:\Users\Administrator\Desktop\ProjectDataDashboard\visualizations\viz2_top_countries.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print(f"\n INSIGHT: {country_counts.index[0]} dominates with {country_counts.iloc[0]:,} titles")
print(f"Top 3 countries: {country_counts.index[0]}, {country_counts.index[1]}, {country_counts.index[2]}")
print(f"Top 3 represent {country_counts.head(3).sum():,} titles ({country_counts.head(3).sum()/len(all_countries)*100:.1f}% of all content)")
```

## Content Addition Trend Over Years

```python
df['year_added'] = df['date_added'].dt.year
yearly_counts = df['year_added'].value_counts().sort_index().dropna()
```

```python
plt.figure(figsize = (14, 7))
plt.plot(yearly_counts.index, yearly_counts.values, marker = 'o', color = '#E50914', linewidth = 3, markersize = 8)
plt.fill_between(yearly_counts.index, yearly_counts.values, alpha=0.3, color='#E50914')
plt.title('Netflix Content Addition Trend Over Years', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=12, fontweight='bold')
plt.ylabel('Number of Titles Added', fontsize=12, fontweight='bold')
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True, alpha = 0.3, linestyle = '--')
peak_year = yearly_counts.idxmax()
peak_value = yearly_counts.max()
plt.annotate(f'Peak: {int(peak_value)} titles', xy = (peak_year, peak_value), xytext = (peak_year - 1, peak_value + 200), arrowprops = dict(arrowstyle = '->' , color = 'black', lw = 2), fontsize = 12, fontweight = 12)
plt.tight_layout()
plt.savefig(r'C:\Users\Administrator\Desktop\ProjectDataDashboard\visualizations\viz3_yearly_trend.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print(f"\n INSIGHT: Peak content addition in {int(peak_year)} with {int(peak_value):,} titles")
recent_growth = ((yearly_counts.iloc[-1] - yearly_counts.iloc[-3]) / yearly_counts.iloc[-3]) * 100
print(f"Recent trend (last 2 years): {recent_growth:+.1f}% change")
```

```python
top_rating = df['rating'].value_counts().index[0]
top_rating_count = df['rating'].value_counts().iloc[0]
print(f"\n INSIGHT: '{top_rating}' is most common with {top_rating_count:,} titles")
print(f"This suggests Netflix targets primarily {'families/general audience' if 'PG' in top_rating else 'mature audiences'}")
```

## Top 10 Genres

```python
genres_split = df['listed_in'].dropna().str.split()
all_genres = []
for genre in genres_split:
    all_genres.extend(genre)
genre_counts = pd.Series(all_genres).value_counts().head(10)
```

```python
plt.figure(figsize=(12, 7))
genre_counts.plot(kind='barh', color='#E50914', edgecolor='black', linewidth=1.2)
plt.title('Top 10 Most Popular Genres on Netflix', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Number of Titles', fontsize=12, fontweight='bold')
plt.ylabel('Genre', fontsize=12, fontweight='bold')
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
for i, v in enumerate(genre_counts):
    plt.text(v + 20, i, str(v), va='center', fontweight='bold', fontsize=10)
plt.tight_layout()
plt.savefig(r'C:\Users\Administrator\Desktop\ProjectDataDashboard\visualizations\viz5_top_genres.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print(f"\n INSIGHT: '{genre_counts.index[0]}' dominates with {genre_counts.iloc[0]:,} titles")
print(f"Top 3 genres: {', '.join(genre_counts.head(3).index.tolist())}")
```
